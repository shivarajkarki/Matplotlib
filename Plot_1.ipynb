"""

Plot three simple plots

changing color, line and marker

labeling the plots

"""

```python
import numpy as np
import matplotlib.pyplot as plt
```

https://matplotlib.org/3.1.0/api/_as_gen/matplotlib.pyplot.figure.html

```python
plt.rcParams['figure.figsize'] = (4,3) # width 4 inch, height 3 inches
plt.rcParams['figure.dpi'] = 150 # resolution
```

```python
#Basic line plot
xvals = np.arange(0,10, 0.1)
plt.plot(xvals, np.sin(xvals))
```

```python
# label axis
plt.plot(xvals, np.sin(xvals))
plt.xlabel('x axis')
plt.ylabel('y axis')
```

```python
#multiple plots
plt.plot(xvals, np.sin(xvals))
plt.plot(xvals, np.cos(xvals))
```

```python
# legend
plt.plot(xvals, np.sin(xvals), label = '$y = \sin(x)$')
plt.plot(xvals, np.cos(xvals), label = '$y = \cos(x)$')
plt.legend()
plt.xlabel('$x$')
plt.ylabel('$y$')
```

```python
# legend location, title, color
plt.plot(xvals, np.sin(xvals), label = '$y = \sin(x)$', color = 'red')
plt.plot(xvals, np.cos(xvals), label = '$y = \cos(x)$', color = 'green')
plt.legend(loc = 'upper right')
plt.xlabel('$x$')
plt.ylabel("$y$")
plt.title('Sinusoida')
```

```python
#marker linestyle
#linestyle available {:, -., --, -}
plt.plot(xvals, np.sin(xvals), label = '$y = \sin(x)$', color = 'red', linestyle = '--')
plt.plot(xvals, np.cos(xvals), label = '$y = \cos(x)$', marker = 'o', linestyle = 'none')
plt.legend(loc = 'upper right')
plt.xlabel('$x$')
plt.ylabel("$y$")
plt.title('Sinusoida')
```

```python
#short form of color and marker
#   b : blue.
#   g : green.
#   r : red.
#   c : cyan.
#   m : magenta.
#   y : yellow.
#   k : black.
#   w : white.

plt.plot(xvals, np.sin(xvals), 'r+--',label = '$y = \sin(x)$')
plt.plot(xvals, np.cos(xvals), 'k-.',label = '$y = \cos(x)$')
plt.legend(loc = 'upper right')
plt.xlabel('$x$')
plt.ylabel("$y$")
plt.title('Sinusoida')
```

```python
#axis limit
plt.plot(xvals, np.sin(xvals), 'r--',label = '$y = \sin(x)$')
plt.plot(xvals, np.cos(xvals), 'bo-',label = '$y = \cos(x)$')
plt.legend(loc = 'upper right')

plt.xlim(0, 2)
plt.ylim(0, 0.5)

plt.xlabel('$x$')
plt.ylabel("$y$")
plt.title('Sinusoida')
```
