```python
#basic Scatter Plot

import numpy as np
import matplotlib.pyplot as plt

xvals = np.arange(0,10,0.1)
plt.scatter(xvals, np.sin(xvals))
```

```python
#color = c
plt.scatter(xvals, np.sin(xvals), c = 'r')
```

```python
plt.scatter(xvals, np.sin(xvals), c = np.sin(xvals))
plt.colorbar()
```

```python
#size = s
plt.scatter(xvals, np.sin(xvals), c = np.sin(xvals), s = np.power(xvals, 3))
```

```python
#make outer edge disappear
plt.scatter(xvals, np.sin(xvals), c = np.sin(xvals), s = np.power(xvals, 3), edgecolor = 'none')
```

```python
plt.scatter(xvals, np.sin(xvals), c = np.sin(xvals), s = np.power(xvals, 3), edgecolor = 'none', cmap = 'inferno')
```
