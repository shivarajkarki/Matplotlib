# bar()

```python
#load numpy for math/array operations
#load matplotlib for plotting
import numpy as np
import matplotlib.pyplot as plt
```

https://kite.com/python/docs/matplotlib.rcParams

```python
#setup figure size and DPI for screen demo
plt.rcParams['figure.figsize'] = (4,3)
plt.rcParams['figure.dpi'] = 150

#plt.figure(figsize = (,), dpi)
```

```python
# basic bar plot
nums = np.random.uniform(size = 10)
```

```python
nums
```

```python
#for bar plot first need to send positional argument i,e x axis
plt.bar(np.arange(10), height = nums)
```

```python
# horizontal bar plot
#barh
plt.barh(np.arange(10), width=nums)
```

```python
#colors
plt.barh(np.arange(10), width=nums, color = 'red')
```

```python
#edgecolor
plt.barh(np.arange(10), width=nums, color = 'yellow', edgecolor = 'r')
```

```python
#alignment
plt.bar(np.arange(10), height=nums, align = 'center')
```

```python
plt.bar(np.arange(10), height=nums, align = 'edge')
```

```python
#hatch and fill
plt.bar(np.arange(10), nums, hatch = 'x')
```

```python
#hatch and fill
plt.bar(np.arange(10), nums, hatch = '|')
```

```python
 #hatch and fill
plt.bar(np.arange(10), nums, hatch = '-')
```

```python
#width
plt.bar(np.arange(10), nums, width = 1)
```

```python
#width
plt.bar(np.arange(10), nums, width = 0.5)
```

```python
#bottom alignment

plt.bar(np.arange(10), nums, bottom=10)
```

```python
plt.bar(np.arange(10), nums, bottom=0.1*nums)
```

# hist()

```python
#basic histogram
rands = np.random.normal(size = int(1e6))
plt.hist(rands)
#displays tuple
#first array takes bin height
#second is bin start point
```

```python
#basic histogram
#supress array display use semicolon
rands = np.random.normal(size = int(1e6))
plt.hist(rands);
```

```python
#limit the display range
plt.hist(rands, range = (0,6))
```

```python
#limit the display range
plt.hist(rands, range = (0,1))
```

```python
#chose no of bins

plt.hist(rands, bins = [0,1,2,3])
```

```python
plt.hist(rands, bins = 'auto');
```

```python
plt.hist(rands, bins = 'auto', histtype='stepfilled');
```

```python
#Display cumlative distribution of gaussian
plt.hist(rands, bins = 'auto', cumulative=True);
```

```python
#to get normalised CDF
plt.hist(rands, bins = 'auto', cumulative=True, normed=True);
```

```python
#multiple histogram
plt.hist(rands, bins = 'auto');
plt.hist(0.5*rands, bins = 'auto');
```

```python
#multiple histogram
#send tuple
plt.hist((rands, 0.5*rands), bins = 'auto', stacked=True);
```
