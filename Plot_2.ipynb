Scatter() and plot()

```python
import numpy as np
import matplotlib.pyplot as plt

plt.rcParams['figure.figsize'] = (4,3) # width 4 inch, height 3 inches
plt.rcParams['figure.dpi'] = 150 # resolution

xvals = np.arange(0,10, 0.1)

plt.plot(xvals, np.sin(xvals), label = '$y = \sin(x)$')
plt.plot(xvals, np.cos(xvals), label = '$y = \cos(x)$')
plt.legend(loc = 1)
plt.xlabel('$x$')
plt.ylabel('$y$')
```

```python
#join() method is a string method and returns a string in which the elements of sequence have been joined by str separator.
idx = 0
for marker in ('-', '-.', '--', '-'):
    plt.plot(xvals, np.sin(xvals)+idx, ''.join(('k',marker)))
    idx += 1
```

```python
xvals2 = np.arange(0,10, 0.01)

plt.plot(xvals2, np.sin(xvals2), 'k.')
```

```python
#use markevery to mark every 10 points not all
#use markersize to set size of marker
xvals2 = np.arange(0,10, 0.01)

plt.plot(xvals2, np.sin(xvals2), 'k.' ,markevery = 10, markersize = 4)
```

```python
#use markevery to mark every 10 points not all
#use markersize to set size of marker
xvals2 = np.arange(0,10, 0.01)

plt.plot(xvals2, np.sin(xvals2), 'r.' ,markevery = 10, markersize = 1)
```

```python
#marker can be defined

plt.plot(xvals, np.sin(xvals), marker = r'$\Sigma$' ,markevery = 2, markersize = 10)
```

```python
#marker can be defined

plt.plot(xvals, np.sin(xvals), marker = r'$\Sigma$' ,linestyle = 'none',markevery = 3, markersize = 10)
```

```python
#linewidth = lw
#zorder can make plot appear front or back i,e z-axis

plt.plot(xvals, np.sin(xvals), lw = 3, zorder = 10)
plt.plot(xvals+1, np.sin(xvals), lw = 8, zorder = 1)
```

```python
#transparency

plt.plot(xvals, np.sin(xvals), lw = 8, alpha = 0.5)
plt.plot(xvals+1, np.sin(xvals), lw = 8, alpha = 0.5)
```
