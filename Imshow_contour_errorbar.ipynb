# Imshow

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize = (4,3), dpi = 150)
```

```python
#generate smoothed, gaussian random field
from scipy.ndimage.filters import gaussian_filter
rands2d = gaussian_filter(np.random.normal(size=(512,512)), sigma = 10)
```

```python
#basic imshow
plt.imshow(rands2d)
plt.colorbar()
```

```python
#extent (Left, Right, Bottom, Top)
#limit are described 
plt.imshow(rands2d, extent = [-1,1,-1,1])
```

```python
#limit the colorbar 
plt.imshow(rands2d, vmax = 0.5, vmin = -0.5)
plt.colorbar()
```

```python
#aspect ratio < 1 horizontal
#aspect ratio > 1 vertical

plt.imshow(rands2d, aspect=0.5)
```

```python
plt.imshow(rands2d, aspect=2)
```

```python
#interpolation
small = np.arange(0,4).reshape(2,2)
plt.imshow(small)
```

```python
#interpolation
small = np.arange(0,4).reshape(2,2)
plt.imshow(small, interpolation='gaussian')
```

```python
#interpolation
small = np.arange(0,4).reshape(2,2)
plt.matshow(small)
```

# Contours

https://en.wikipedia.org/wiki/Contour_line

```python
#basic Contours
#an outline represention
srands2 = gaussian_filter(rands2d, sigma = 100)
plt.contour(srands2)
plt.colorbar()
```

```python
#contour levels
#select the contour levels
plt.contour(srands2, levels = [0, 0.003])
```

```python
#specify no. of contour lines
plt.contour(srands2, 30)
```

```python
plt.contour(srands2, 50)
```

```python
#filled contour
plt.contourf(srands2,10)
```

```python
plt.contourf(srands2, 20)
```

```python
#contour labels
contours = plt.contour(srands2, 5)
plt.clabel(contours)
plt.colorbar()
```

```python
#contour labels
contours = plt.contour(srands2, 15)
plt.clabel(contours)
plt.colorbar()
```

```python
#we are interested in only contour with level = 0, and making it's color as black
plt.contour(srands2, levels=[0], colors = 'k')
```

```python
#overlap
plt.contour(srands2, levels=[0], colors = 'k')
plt.imshow(srands2)
```

# Uncertainities on line and scatter plot

```python
nums = np.arange(0,10,0.5)
xerrors = np.random.uniform(size = 20)
yerrors = np.random.uniform(size = 20)
xerrors
```

```python
#basic error plot
plt.plot(nums, np.sin(nums))
```

```python
#to include error bars to this graph to show uncertainities
#y axis errors
plt.errorbar(nums, np.sin(nums), yerr=yerrors)
```

```python
#x axis errors
plt.errorbar(nums, np.sin(nums), xerr=xerrors)
```

```python
#set color
plt.errorbar(nums, np.sin(nums), xerr=xerrors, ecolor='r')
```

```python
#line width
plt.errorbar(nums, np.sin(nums), xerr=xerrors, elinewidth=0.5)
```

```python
#cap size for error cap
plt.errorbar(nums, np.sin(nums), xerr=xerrors, capsize =10)
```

# Uncertainities on Bar Plots

```python
plt.bar(nums, np.sin(nums)**2, xerr = xerrors)
```

```python
plt.bar(nums, np.sin(nums)**2, yerr = xerrors, capsize = 10, width = 0.5)
```
